## Optimization

#### Create Test Dataframe

```python
import pandas as pd
import numpy as np

df = pd.DataFrame()

df['odds']= money_line['DraftKings'].astype('int')

# Implied Odds
df['PredProb'] = df['odds'].apply(lambda x: -x/(-x+100) if x < 0 else 100/(x+100))

# Randomization
np.random.seed(75) 
df['PredProb'] = df['PredProb']*np.random.uniform(low=0.75,high=1.15,size=df.shape[0])
df['error'] = np.random.uniform(low=0.02,high=0.08,size=df.shape[0])
```

```python
from optimization.optimize_bets import optimize_bets

bets = optimize_bets(df[['PredProb','error']],money_line,1000)
```

```python
bets
```

```python
from optimization.optimize_EV import optimize_EV

bets = optimize_EV(df[['PredProb','error']],money_line,1000)
```
